```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.compose import ColumnTransformer
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Conv1D, MaxPooling1D, Dropout

# Step 1: Data Preparation
data = pd.read_csv("copper_oec_trade_data.csv")

# Extract features and target variable
features = []
target = []

# One-hot encode exporter_name
encoder = LabelEncoder()
exporter_encoded = encoder.fit(data[['exporter_name']])

scaler = StandardScaler()
features_scaled = scaler.fit(data[['export_value','export_quantity']])


# print(len(encoder.classes_))

for year,group in data.groupby('year'):
    arr = np.ones(shape = (len(encoder.classes_),3))*-1    
    num_features = scaler.transform(group[['export_value','export_quantity']])
    cat_features = encoder.transform(group[['exporter_name']])
    
    temp = np.column_stack((num_features,cat_features))
    # print(temp)
    for i in range(len(temp)):
        arr[i] = temp[i]
    target.append(group['Average Closing Price'].mean())
    
    features.append(arr)
    
features = np.array(features)
target = np.array(target)
    
    
# features = features.drop(['exporter_name'], axis=1, inplace=True)

# # Normalize numerical features
# scaler = StandardScaler()
# features_scaled = scaler.fit_transform(features)





# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=50)

print(X_train.shape)
```

```python
# Step 2: Model Definition
model = Sequential()
model.add(Conv1D(32, kernel_size=3, activation='relu', input_shape=(features.shape[1],features.shape[2])))
# model.add(MaxPooling1D(pool_size=2))
model.add(Flatten())
model.add(Dense(16, activation='relu'))
model.add(Dense(1, activation='linear'))  # Linear activation for regression

model.summary()

# Step 3: Compile the Model
model.compile(optimizer='adam', loss='mean_squared_error', metrics=['mean_squared_error'])

# Step 4: Train the Model
model.fit(X_train, y_train, epochs=100, batch_size=10)

# Step 5: Evaluate the Model
loss = model.evaluate(X_test, y_test)
print(f'Mean Squared Error on Test Data: {loss}')

# Make predictions
predictions = model.predict(X_test)
```
